```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress


# Importar os dados
df = pd.read_csv('epa-sea-level.csv')
```

```python
# Criar gráfico de dispersão
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```

```python
from scipy.stats import linregress

# Obter a linha de melhor ajuste
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Criar valores para a linha de melhor ajuste
years = pd.Series(range(1880, 2051))
sea_levels = intercept + slope * years

# Plotar a linha de melhor ajuste
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(years, sea_levels, color='red')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```

```python
# Filtrar os dados a partir do ano 2000
df_recent = df[df['Year'] >= 2000]

# Obter a linha de melhor ajuste para os dados recentes
slope_recent, intercept_recent, r_value_recent, p_value_recent, std_err_recent = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])

# Criar valores para a linha de melhor ajuste recente
years_recent = pd.Series(range(2000, 2051))
sea_levels_recent = intercept_recent + slope_recent * years_recent

# Plotar as linhas de melhor ajuste
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(years, sea_levels, color='red', label='Best Fit Line 1880-2050')
plt.plot(years_recent, sea_levels_recent, color='green', label='Best Fit Line 2000-2050')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.legend()
plt.show()
```

```python
def draw_plot():
    # Importar os dados
    df = pd.read_csv('epa-sea-level.csv')

    # Criar gráfico de dispersão
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
    
    # Obter a linha de melhor ajuste
    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    years = pd.Series(range(1880, 2051))
    sea_levels = intercept + slope * years
    plt.plot(years, sea_levels, color='red', label='Best Fit Line 1880-2050')
    
    # Obter a linha de melhor ajuste para os dados recentes
    df_recent = df[df['Year'] >= 2000]
    slope_recent, intercept_recent, r_value_recent, p_value_recent, std_err_recent = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
    years_recent = pd.Series(range(2000, 2051))
    sea_levels_recent = intercept_recent + slope_recent * years_recent
    plt.plot(years_recent, sea_levels_recent, color='green', label='Best Fit Line 2000-2050')
    
    # Configurar o gráfico
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.legend()
    
    # Salvar a imagem
    plt.savefig('sea_level_plot.png')
    return plt.gca()

# Chamar a função para gerar e salvar a imagem
draw_plot()
```
